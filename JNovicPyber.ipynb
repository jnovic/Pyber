1.) Urban residents pay a lower average fair (per city) while their are more rides (per city), while Rural residence pay more on average but take fewer rides (per city).
2.) Most fares (62.7%) come from urban riders, while suburban riders contribute a substantial amount (30.5%) to the total fares.
3.) Most drivers (80.9%) work in urban environments.

```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_filepath = "data/city_data.csv"
ride_filepath = "data/ride_data.csv"

city_df = pd.read_csv(city_filepath)
ride_df= pd.read_csv(ride_filepath)
```

```python
merge_df = pd.merge(city_df, ride_df, on="city")

merge_df = merge_df.set_index("type")
merge_df.head()
```

# Bubble Plot

```python
test_df = merge_df.loc["Urban"]
test_df2 = merge_df.loc["Suburban"]
test_df3 = merge_df.loc["Rural"]
```

```python
#Average Fare Per City
groupby_urban = test_df.groupby(["city"])
groupby_sub = test_df2.groupby(["city"])
groupby_rural = test_df3.groupby(["city"])
```

```python
ave_fareurban = groupby_urban["fare"].mean()
ave_faresub = groupby_sub["fare"].mean()
ave_farerural = groupby_rural["fare"].mean()
```

```python
#Ride per city
ride_city = groupby_urban["ride_id"].count()
ride_city_sub = groupby_sub["ride_id"].count()
ride_city_rural = groupby_rural["ride_id"].count()
```

```python
#num of drivers per city
drivers_city = merge_df.iloc[:, 0:2]
drivers_city = drivers_city.sort_values("city")
drivers1 = drivers_city.loc["Urban"]
drivers1= drivers1["driver_count"].unique()
drivers2 = drivers_city.loc["Suburban"]
drivers2= drivers2["driver_count"].unique()
drivers3 = drivers_city.loc["Rural"]
drivers3= drivers3["driver_count"].unique()
```

```python
scatter1 = plt.scatter(ave_fareurban, ride_city, s=drivers1 , c="skyblue", edgecolors="black")
scatter2 =plt.scatter(ave_faresub, ride_city_sub, s=drivers2, c="gold", edgecolors="black")
scatter3 =plt.scatter(ave_farerural, ride_city_rural, s=drivers3, c="coral", edgecolors="black")
plt.title("Pyber Ride Share Data (2016)")
plt.xlabel("Average Fare (Per City)")
plt.ylabel("Total Number of Rides (Per City)")
plt.legend(handles=[scatter1, scatter2, scatter3], labels= ["Urban", "Suburban", "Rural"],loc="best")
```

```python
#pie chart

sizes = [6.81 , 30.46, 62.72]

# The colors of each section of the pie chart
colors = ["gold", "lightcoral", "lightskyblue"]

# Tells matplotlib to seperate the "Python" section from the others
explode = (0, 0, 0.1)

label = ["Rural", "Suburban", "Urban"]
# The colors of each section of the pie chart



riders_pie =plt.pie(sizes, labels= label, explode=explode,colors=colors, autopct="%1.1f%%", shadow=True, startangle=140)
plt.axis("equal")
plt.title("Percentage of Total Fares ")
print(riders_pie)
```
